```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('../../datasets/titanic_toy.csv')
```

```python
data.head(4)
```

```python
x = data.drop(columns=['Survived','Family'])
y = data['Survived']
```

```python
x.head(5)
```

```python
y.head(5)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

Prediction before imputation


```python
from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
si = SimpleImputer()
x_train_trf = si.fit_transform(x_train)
x_test_trf = si.fit_transform(x_test)
```

```python
x_train_trf
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(x_train_trf,y_train)
y_pred = clf.predict(x_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
