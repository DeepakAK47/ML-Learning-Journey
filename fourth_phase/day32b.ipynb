# Calculating accuracy score after applying binarization

```python
# binarisation convert the data between 0 and 1
from sklearn.preprocessing import Binarizer
```

```python
data.head(1)
```

```python
# importing column transformer
from sklearn.compose import ColumnTransformer
```

```python
# applying Binarization over family column
trf = ColumnTransformer([
    ('trf',Binarizer(copy=False),['Family']) # copy=False -> It means do not make copy of data just modify inplace data
],remainder='passthrough')
```

```python
data.head(1)
```

```python
x_train.head(1)
```

```python
# transforming x_train and  _x_test data using the binarizatiom
x_train_trf = trf.fit_transform(x_train) # fitting is already done here
x_test_trf = trf.fit_transform(x_test)
```

```python
# Note -> Always be remember after applying the transformation data converted in numpy array
```

```python
# converting transform data from numpy to dataframe
x_train_trf = pd.DataFrame(x_train_trf,columns=['Family','Age','Fare'])
```

```python
x_train_trf.head(1)
```

```python
# Note -> sklearn do not care which kind of data you are passing dataframe or numpy array during fitting model.
```

```python
clf.fit(x_train_trf,y_train)
y_pred2 = clf.predict(x_test_trf)
print('The accuracy score of the model after applying the binarization : ',accuracy_score(y_pred2,y_test))
```
