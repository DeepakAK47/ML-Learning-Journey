# Working With date

```python
# converting date into datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
date.head()
```

```python
# Extracting year - > Making a new column for year
date['year'] = date['date'].dt.year
date.head(2)
```

```python
# Extracting month
date['month'] = date['date'].dt.month
date.head()
```

```python
# Extracting date of month
date['month_name'] = date['date'].dt.month_name()
date.head()
```

```python
# Extracting month date
date['month_date'] = date['date'].dt.day
date.head()
```

```python
// Doubt
```

```python
# Extracting day of week
date['week'] = date['date'].dt.dayofweek
date.head(20)
```

```python
# Extracting day of that day
date['day'] = date['date'].dt.day_name()
date.head(3)
```

```python
# dropping unnecessary columns
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is Weekend
date['is_weekend'] = np.where(date['day'].isin(['Sunday','Saturday']),1,0)
date.head()
```

```python
# Extrating week of year
date['week_year'] = date['date'].dt.isocalendar().week   
date.head()
```

```python
# Extracting quater
date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
#extracting the semester
date['semester'] = np.where(date)
```
