# EDA(Exploratory Data Analysis) Univariate Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('titanic.csv')
```

```python
data
```

# 1. Categorical Data

# Countplot

```python
# Note -> Here 0 for male ans 1 for female
sns.countplot(x='Sex', data=data)
```

```python
data['Sex'].value_counts()
```

```python
sns.countplot(x='Pclass',data=data)
```

```python
sns.countplot(x='2urvived',data=data)
```

# PieChart

```python
# Note -> We can draw pie chart using seaborn only for this we have to use matplotlib
data['Sex'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
data['Pclass'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
data['2urvived'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```
