# feature construction and feature splitting

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../../datasets/titanic2.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
data.head()
```

```python
# dropping missing values
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
x = data.drop(columns=['Survived'])
y = data['Survived']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
cv = cross_val_score
lr = LogisticRegression()
```

```python
# accuracy_score
np.mean(cross_val_score(lr,x,y,scoring='accuracy',cv=20))
```

# Applying Feature construction

```python
x['family_size'] = x['SibSp'] + x['Parch'] + 1
```

```python
x.head()
```

```python
# defining family size
def myfun(num):
    if(num == 1): # alone
        return 0
    elif (num > 2 & num < 4): # small family
        return 1
    else:
        return 2 # large_family
    
```

```python
x['family_type'] =  x['family_size'].apply(myfun)
```

```python
x.head()
```

```python
x.drop(columns=['SibSp','Parch','family_size'])
```

```python
# prediction -> After feature construction
np.mean(cross_val_score(lr,x,y,scoring='accuracy',cv=20))
```
