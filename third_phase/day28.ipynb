```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
### splitting data into x and y
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data.drop(columns=['has_covid']),data['has_covid'],test_size=0.3,random_state=42)
```

```python
x_train.shape[0]
```

# Aam Zindagi

```python
# Adding simple inputer to the fever column
si = SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])

# Applying simple imputer over the test data also
x_test_fever = si.fit_transform(x_test[['fever']])
```

```python
x_train_fever
```

```python
# Applying Ordinal Encoding to cough column
oe = OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])

# Applying Ordinal Encoding over test data 
x_test_cough = oe.fit_transform(x_test[['cough']])
```

```python
x_train
```

```python
x_train_cough
```

```python
# Applying one hot encoding over gender and city
ohe = OneHotEncoder(drop='first',sparse_output=False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])

# Aplying one hot encoding over test data
x_test_gender_city = ohe.fit_transform(x_test[['gender','city']])
```

```python
# Extracting the age column from the data 
x_train_age = x_train.drop(columns=['gender','fever','cough','city']).values
x_test_age = x_test.drop(columns=['gender','fever','cough','city']).values 
```

```python
# Merging all the column into single one
x_train_transform = np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis=1)
x_test_transform = np.concatenate((x_test_age,x_test_fever,x_test_gender_city,x_test_cough),axis=1)
```

# Mentoss Zindagi

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tf1',SimpleImputer(),['fever']),
    ('tf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(x_train)
```

```python
transformer.fit_transform(x_test)
```
