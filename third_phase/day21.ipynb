# Applying bivariate analysis on (Numeriacal - Numerical data type columns)

```python
data.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data,hue='sex')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data,hue='sex',style='smoker')
```

### What is size signifises here


```python
sns.scatterplot(x='total_bill',y='tip',data=data,hue='sex',style='smoker',size=data['size'])
```

# 5. HeatMap (Categorical - Categorical)

```python
# determining number of survival in each classes
pd.crosstab(data2['Pclass'],data2['2urvived'])
```

```python
pd.crosstab(data2['Sex'],data2['2urvived'])
```

### Note -> pd.crosstab shows how many times combinations of two categorical variables occur.

```python
pd.crosstab(data2['Pclass'],data2['Sex'])
```
