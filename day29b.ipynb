### testing the train model of titanic datasets without using teh pipelines


```python
import numpy as np
import pickle
```

```python
ohe_sex = pickle.load(open('models/ohe_sex.pkl','rb'))
ohe_embarked = pickle.load(open('models/ohe_embarked.pkl','rb'))
clf = pickle.load(open('models/clf.pkl','rb'))
```

```python
# taking a random test data as 
# Pclass/gender/age/sibsp/Parch/Embarked
test_input = np.array([2,'male',31.0,0,0,'S'],dtype=object).reshape(1,6)
```

```python
test_input
```

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1))
```

```python
test_input_sex
```

```python
test_input_embarked = ohe_embarked.transform(test_input[:,-1].reshape(1,1))
```

```python
test_input_embarked
```

```python
test_input_age = test_input[:,2].reshape(1,1)
```

```python
test_input_age
```

```python
# concatiatinf all modified columns with the remaining columns
test_input_transformed = np.concatenate((test_input[:,[0,3,4]],test_input_age,test_input_sex,test_input_embarked),axis=1)
```

```python
test_input_transformed.shape[1]
```

```python
# predicting the out whether the person will survive or not according to the trained data
clf.predict(test_input_transformed)
```
